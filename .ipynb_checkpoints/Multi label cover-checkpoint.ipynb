```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from fastai.vision.all import *
from fastai.text.all import *
from fastai.collab import *
from fastai.tabular.all import *
```

```python
import os
os.getcwd()
%cd "UCMerced"
```

```python
df = pd.read_csv(r"D:\python_projects\Multi-label Cover\UCMerced\multilabels.csv", sep="\t")
df.head()
```

```python
path = Path (r"D:\python_projects\Multi-label Cover\UCMerced\images")
files =get_image_files(path)
len(files)
```

```python
tfms = aug_transforms(flip_vert=True, max_lighting=0.1, max_zoom=1.05, max_warp=0, size=256,
                     xtra_tfms = Normalize.from_stats(*imagenet_stats) )
```

```python
dls = ImageDataLoaders.from_df(df, path, valid_col='is_valid', label_delim=' ', item_tfms=Resize(460),
                              batch_tfms = tfms)
```

```python
dls.show_batch()
```

```python
learn = vision_learner(dls, models.resnet34, metrics=accuracy)
```

```python
learn.lr_find()
```

```python
learn.freeze()
lr = 1e-3
```

```python
learn.fit_one_cycle(5, slice(lr))
```
